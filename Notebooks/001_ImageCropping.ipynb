```python
import numpy as np 
import cv2 
from glob import glob
import matplotlib.pyplot as plt
import os
```

```python
mpath = glob('./Data/Train/Male/*.jpg')[:5000]
fpath = glob('./Data/Train/Female/*.jpg')[:5000]
```

```python
print('Number of images of Female = ',len(fpath))
print('Number of images of Male  = ',len(mpath))
```

```python
img = cv2.imread(fpath[0]) #cv2 read images in BGR 
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # convert image to RGB

haarcas = cv2.CascadeClassifier('./Model/haarcascade_frontalface_default.xml')
img_gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
faces_list = haarcas.detectMultiScale(img_gray,1.3,5)
print(faces_list)

for x,y,w,h in faces_list:
    
    roi = img_rgb[y:y+h,x:x+w] # crop the face out of the image
    
    plt.imshow(roi)
    plt.axis('off')
    plt.show()
```

```python
# save cropped image in a new folder 
for i in range(len(fpath)):
    try:
        img = cv2.imread(fpath[i]) #cv2 read images in BGR 
        img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # convert image to RGB

        img_gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
        faces_list = haarcas.detectMultiScale(img_gray,1.3,5)

        for x,y,w,h in faces_list:
            roi = img[y:y+h,x:x+w] # crop the face out of the image
            cv2.imwrite(f'./Cropped_data/Train/Female/Female_{i}.jpg', roi)
            
            print('Image is processed and saved successfully')
        
    except:
        print('Failed to process the image.')
```

```python
# save cropped image in a new folder 
for i in range(len(mpath)):
    try:
        img = cv2.imread(mpath[i]) #cv2 read images in BGR 
        img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # convert image to RGB

        img_gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
        faces_list = haarcas.detectMultiScale(img_gray,1.3,5)

        for x,y,w,h in faces_list:
            roi = img[y:y+h,x:x+w] # crop the face out of the image
            cv2.imwrite(f'./Cropped_data/Train/Male/Male_{i}.jpg', roi)
            
            print('Image is processed and saved successfully')
        
    except:
        print('Failed to process the image.')
```

```python
# Cropped the images in testing set
mpath_test = glob('./Data/Test/Male/*.jpg')[:1250]
fpath_test = glob('./Data/Test/Female/*.jpg')[:1250]

mpath_val = glob('./Data/Validation/Male/*.jpg')[:1250]
fpath_val = glob('./Data/Validation/Female/*.jpg')[:1250]

dict1 = {'Test_Male': mpath_test, 'Test_Female': fpath_test, 'Validation_Male': mpath_val, 'Validation_Female': fpath_val}

for key,path in dict1.items():
    print(key,len(path))
```

```python
for key,path in dict1.items():
    
    data_set = key[:key.index('_')]
    gender = key[key.index('_')+1:]
    
    for i in range(len(path)): 
        try:
            img = cv2.imread(path[i]) #cv2 read images in BGR 
            img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # convert image to RGB

            img_gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
            faces_list = haarcas.detectMultiScale(img_gray,1.3,5)

            for x,y,w,h in faces_list:
                roi = img[y:y+h,x:x+w] # crop the face out of the image
                cv2.imwrite(f'./Cropped_data/{data_set}/{gender}/{gender}_{i}.jpg', roi)

                print('Image is processed and saved successfully')

        except:
            print('Failed to process the image.')
    
```
